## **Tóm tắt dự án**

#### Những phát hiện chính:

* **Sự nhất quán giữa Energy và Loudness:** Phân tích hồi quy cho thấy một mối tương quan thuận mạnh mẽ ($r = 0.76$) giữa năng lượng và âm lượng. Điều này chứng minh rằng trong sản xuất âm nhạc tiêu chuẩn, cường độ âm thanh là yếu tố cốt lõi tạo nên cảm giác sôi động cho người nghe.

* **Khám phá bất ngờ về "Nghịch lý âm thanh":** Phát hiện thú vị nhất là sự tồn tại của một nhóm ngoại lệ có **Energy cực cao (≈ 1.0) nhưng Loudness cực thấp (< -15dB)**. Qua kiểm chứng, nhóm này không phải âm nhạc thông thường mà là **Âm thanh chức năng (Functional Audio)** như tiếng mưa, tiếng sấm hoặc tiếng ồn trắng. Thuật toán Spotify đánh giá Energy dựa trên mật độ tần số, khiến các âm thanh môi trường dày đặc bị hiểu nhầm là có năng lượng cao như nhạc Rock hay EDM.

* **Công thức để tạo bài Hit:** Phân tích tương quan và biểu đồ Radar cho thấy các bài hát Hit có xu hướng Sôi động và Rõ lời.

	* **Dương tính:** Các yếu tố Loudness, Danceability, và Energy có mối tương quan dương với độ phổ biến. Khán giả đại chúng ưu tiên âm nhạc kích thích thính giác mạnh và phù hợp để chuyển động.

	* **Âm tính:** Yếu tố Instrumentalness có tương quan âm mạnh nhất. Nhạc thuần không lời rất khó trở thành Hit đại chúng so với nhạc có giọng hát.

	* Ngoài ra, các thể loại nhạc đại chúng như các dòng nhạc Pop, Chill, Sad,… giúp tăng tỷ lệ trở thành hit cao hơn

* **Về giọng điệu (major và minor):**

	* **Cảm xúc (valence):** Giọng Trưởng (major) mang lại cảm giác tích cực/vui vẻ hơn
	giọng Thứ (minor), tuy nhiên mức chênh lệch là rất nhỏ. Điều này cho thấy trong
	nhạc hiện đại, điệu thức không còn là yếu tố duy nhất quyết định cảm xúc bài hát.

	* **Năng lượng (energy):** Trái với suy nghĩ truyền thống rằng giọng Thứ thường u
	buồn, yếu đuối, kết quả cho thấy nhạc giọng Thứ có mức năng lượng trung bình
	cao hơn giọng Trưởng. Điều này phản ánh sự trỗi dậy của các dòng nhạc điện tử
	(EDM, Techno) thường sử dụng giọng Thứ để tạo sự bùng nổ

* **Về nhạc nhạy cảm (explicit và non-explicit):**

	* **Định hình phong cách:** Nhạc explicit không chỉ khác biệt ở ca từ mà còn sở hữu
	một cấu trúc âm thanh đặc trưng: ồn ào hơn và nhiều lời hơn.

	* **Khả năng tương tác:** Nhạc **explicit** có chỉ số **danceability** (độ dễ nhảy) cao hơn
	đáng kể so với nhạc **non-explicit**. Điều này bác bỏ định kiến rằng nhạc chứa từ
	ngữ mạnh thì khó tiếp cận; ngược lại, chúng rất phù hợp cho các hoạt động vận
	động mạnh và giải trí tại các câu lạc bộ


#### Giới hạn:

* **Thiên kiến nền tảng:** Dữ liệu được thu thập từ Spotify, do đó nó chỉ phản ánh thị
hiếu của người dùng trên nền tảng này (thường là giới trẻ, hoặc những người rành
công nghệ). Nó có thể không đại diện cho toàn bộ thị trường âm nhạc toàn cầu.

* **Tính chất thuật toán:** Các chỉ số như **valence, danceability, energy** là do thuật
toán của Spotify tự động chấm điểm. Đây là các chỉ số kỹ thuật, đôi khi không
phản ánh chính xác 100% cảm nhận chủ quan của con người (ví dụ: một bài hát có
giai điệu buồn nhưng beat mạnh có thể bị thuật toán chấm điểm **valence** thấp
nhưng **energy** cao gây hiểu nhầm)

* **Hạn chế về dữ liệu:** Tập dữ liệu không phân loại rõ ràng giữa "âm nhạc" (Music) và "âm thanh môi trường" (Nature sounds), dẫn đến việc các chỉ số thống kê bị nhiễu nếu không thực hiện các bước lọc ngoại lệ thủ công.

* **Hạn chế về phạm vi:** Đồ án chưa thể đi sâu vào phân tích sự thay đổi của các chỉ số này theo thời gian (năm phát hành) để thấy được xu hướng sản xuất âm nhạc hiện đại.

* **Mô hình học máy tốt nhất (XGBoost/Random Forest):** chỉ giải thích được khoảng 63% (R2 ~ 0.63) sự biến thiên về độ phổ biến của bài hát.

	* Các yếu tố kỹ thuật (âm thanh, nhịp điệu, cấu trúc), thể loại và tác giả chỉ đóng góp hơn một nửa vào sự thành công. Gần 40% còn lại phụ thuộc vào các yếu tố ngoại sinh không có trong dữ liệu như: ngân sách Marketing, xu hướng mạng xã hội và thời điểm phát hành.

* **Tương quan không phải là Nhân quả:** Kết quả cho thấy nhạc **explicit** có độ ồn
lớn hơn, nhưng chúng ta không thể khẳng định rằng vì bài hát ồn nên nó bị gán
nhãn explicit. Mối quan hệ này chỉ là tương quan đi kèm.

* **Chưa phân tích sâu theo thể loại:** Chúng ta đã gộp chung toàn bộ 114,000 bài
hát để so sánh. Thực tế, sự khác biệt giữa major/minor có thể rất khác nhau trong
nhạc Jazz so với nhạc Metal.

* **Chưa phân tích nội dung lời bài hát:** Đối với câu hỏi về nhạc **explicit**, chúng ta
chỉ dựa vào các chỉ số âm thanh kỹ thuật. Dự án chưa thực hiện Xử lý ngôn ngữ tự
nhiên (NLP) để phân tích mức độ thô tục hay chủ đề của lời bài hát

#### Định hướng tương lai:

* **Phân tích sâu hơn:** Áp dụng các mô hình Machine Learning (như Random Forest) để dự đoán thể loại nhạc dựa trên các đặc tính âm học.

* **Phương pháp thay thế:** Sử dụng kỹ thuật **Clustering (Phân cụm)** để tự động tách biệt nhóm nhạc thông thường và nhóm âm thanh chức năng thay vì lọc thủ công bằng hành lang tin cậy.

* **Dữ liệu bổ sung:** 

	* **Dữ liệu thời gian:** Thu thập thêm dữ liệu về xu hướng âm nhạc theo thời gian để phân tích sự thay đổi gu âm nhạc của người nghe và các yếu tố ảnh hưởng đến độ phổ biến của bài hát.

	* **Lời bài hát:** Để áp dụng NLP (Xử lý ngôn ngữ tự nhiên) phân tích cảm xúc của ca
	từ, so sánh với cảm xúc của giai điệu

	* **Dữ liệu nhân khẩu học:** Độ tuổi và quốc gia của người nghe để tối ưu hóa gợi ý
	theo vùng miền.

* **Phân tích đa biến:** Thay vì kiểm định từng cặp biến đơn lẻ (T-test), sử dụng PCA
(Principal Component Analysis) để giảm chiều dữ liệu và tìm ra các nhóm bài hát.

* **Dashboard tương tác:** Xây dựng ứng dụng web (bằng Streamlit hoặc Tableau)
cho phép người dùng tự tải nhạc lên và xem biểu đồ phân tích bài hát của chính
họ. 

#### Cảm nhận cá nhân (Vũ Hoàng Minh):

* **Thử thách và khó khăn:** 

	* **Sự mâu thuẫn giữa lý thuyết âm nhạc và dữ liệu thực tế:** Một khó khăn lớn là việc giải thích các "nghịch lý" trong dữ liệu. Ví dụ, tại sao Mode Thứ (Minor) thường được coi là buồn nhưng lại có độ Danceability cao hơn Mode Trưởng (Major), hay tại sao tiếng mưa rơi lại có Energy tối đa. Điều này đòi hỏi tôi phải tạm gác lại những định kiến chủ quan để nghiên cứu sâu vào cách thuật toán Spotify định nghĩa các chỉ số này.

	* **Cách tôi đã vượt qua:** Tôi đã áp dụng quy trình kiểm chứng chéo (Cross-verification): mỗi khi có kết quả bất thường, tôi không chỉ nhìn vào biểu đồ mà còn truy xuất trực tiếp các `track_name` cụ thể để nghe thử và đối chiếu. Việc kết hợp giữa **thống kê toán học và cảm thụ âm nhạc thực tế** đã giúp tôi tự tin hơn trong việc đưa ra các nhận xét.

* **Học tập và phát triển:**

	* **Kỹ năng đạt được:** Thành thạo thư viện `statsmodels` để chạy hồi quy OLS, hiểu sâu về **Khoảng dự báo (Prediction Interval)** để xác định ngoại lệ thay vì chỉ dựa vào cảm tính.

	* **Điều gây bất ngờ nhất:** Phát hiện về nhóm âm thanh chức năng có Energy cao nhưng Loudness thấp là một khám phá bất ngờ và thú vị, mở ra hướng nghiên cứu mới về cách các thuật toán đánh giá âm thanh.

	* **Định hình tư duy về Khoa học Dữ liệu:** Dự án này giúp tôi nhận ra rằng Khoa học dữ liệu không chỉ là chạy code hay vẽ biểu đồ đẹp, mà là quá trình tìm ra câu chuyện ẩn sau những con số bất thường. Việc đặt câu hỏi "Tại sao?" khi thấy một điểm dữ liệu nằm ngoài quy luật quan trọng hơn nhiều so với việc chỉ chấp nhận kết quả hiển thị ban đầu.

	* Việc hiểu các thuật ngữ chuyên ngành âm nhạc (Valence, Mode, Key,…) để biện luận kết quả cũng là một thách thức.

	* **Nhận ra Data Science** không chỉ là code và thuật toán, mà quan trọng nhất là các kiến thức liên quan để đặt câu hỏi đúng và hiểu được giới hạn của dữ liệu mình đang có.

	* **Điều bất ngờ nhất** là khi các yếu tố liên quan đến nhạc lý và âm thanh lại ảnh hưởng rất ít đến độ phổ biến của 1 bài hát, thay vào đó là các yếu tố như nghệ sĩ, thể loại nhạc lại ảnh hưởng rất lớn.
